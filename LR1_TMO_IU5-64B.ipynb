```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.datasets import load_iris
sns.set(style="ticks")
```

```python
iris=load_iris()
```

```python
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data.head()
```

Датасет содержит следующие колонки:
sepal length (cm)- длина чашелистика
sepal width (cm)-ширина чашелистика
petal length (cm)- длина лепестка
petal width (cm)- ширина лепестка
целевой признак -target , класс к которому относится растение

```python
mask = np.zeros_like(data.corr(), dtype=np.bool)
# чтобы оставить нижнюю часть матрицы
# mask[np.triu_indices_from(mask)] = True
# чтобы оставить верхнюю часть матрицы
mask[np.tril_indices_from(mask)] = True
sns.heatmap(data.corr(), mask=mask, annot=True, fmt='.3f')
```

```python
fig, ax = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(15,5))
sns.heatmap(data.corr(method='pearson'), ax=ax[0], annot=True, fmt='.2f')
sns.heatmap(data.corr(method='kendall'), ax=ax[1], annot=True, fmt='.2f')
sns.heatmap(data.corr(method='spearman'), ax=ax[2], annot=True, fmt='.2f')
fig.suptitle('Корреляционные матрицы, построенные различными методами')
ax[0].title.set_text('Pearson')
ax[1].title.set_text('Kendall')
ax[2].title.set_text('Spearman')
```

<b>Вывод:Целевой признак наиболее сильно коррелирует с petal length (cm)	и petal width (cm), но также заметна линейная зависимость  и если бы признаков было много ,то можно было бы выкинуть petal length (cm) тк корелляция с целевым меньше незначительно. Одназначно стоит исключить признак sepal width (cm) тк очень плохая отрицательная корелляция с любыми признаками </b>
